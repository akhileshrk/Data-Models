### Hierarchical Clustering 

**Hierarchical Clustering** is a method of cluster analysis which seeks to build a hierarchy of clusters. It works by either iteratively merging smaller clusters into larger ones (**agglomerative method**) or by splitting a large cluster into smaller ones (**divisive method**). This technique does not require the number of clusters to be specified in advance and is often used to explore the dataâ€™s inherent structure.
### Key Applications of Hierarchical Clustering

- **Gene Expression Data Analysis**: Grouping genes with similar expression profiles for biological studies (e.g., cancer research).
- **Customer Segmentation**: Identifying different groups of customers based on purchasing behavior.
- **Document or Text Clustering**: Organizing text documents into related groups (e.g., news articles or academic papers).
- **Image Segmentation**: Grouping image pixels into regions that share similar colors or textures.
- **Social Network Analysis**: Detecting communities or groups within a network of users or entities.
- **Anomaly Detection**: Identifying unusual data points that don't fit well into any cluster.

### First Use in Agriculture: Hierarchical Clustering

- **Crop Variety Classification (1960s, USA)**: The first major applications of hierarchical clustering in agriculture were focused on classifying different crop varieties. Researchers in the 1960s in the USA used clustering techniques to group varieties based on characteristics such as growth rates, yield, and resistance to pests, allowing for better selection and breeding strategies.

- **Soil Classification (1970s, Europe)**: Hierarchical clustering was also applied in Europe during the 1970s for classifying soil types based on their physical and chemical properties. This helped to optimize the use of fertilizers and irrigation based on soil characteristics, leading to more efficient farming practices.

- **Plant Disease Diagnosis (1980s, Japan)**: In Japan in the 1980s, hierarchical clustering was employed to diagnose plant diseases. By grouping similar symptom patterns across crops, researchers could identify new diseases more quickly and develop better disease management strategies.

- **Precision Farming (1990s, Global)**: In the 1990s, the use of hierarchical clustering in precision farming became more widespread. This method was used to analyze spatial patterns of crop yield variability across farms, helping farmers to create tailored management zones for fertilization and irrigation.

- **Genetic Improvement (2000s, Worldwide)**: By the 2000s, hierarchical clustering was widely adopted in genetic research to group plant and animal populations based on their genetic traits. This facilitated better selection in breeding programs to improve crop resistance, yield, and quality.


# Use Case: Hierarchical Clustering for Agricultural Farm Segmentation: Analyzing Farm Characteristics and Crop Patterns

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA

# Set random seed for reproducibility
np.random.seed(42)

# Number of data points (1 Lakh)
n = 1000

# Simulating fixed effects for 15 parameters
temperature = np.random.uniform(18, 30, n)  # Temperature in Celsius
humidity = np.random.uniform(40, 80, n)  # Humidity in percentage
soil_ph = np.random.uniform(5.5, 7.5, n)  # Soil pH
fertilizer = np.random.choice([0, 1], size=n)  # 0 = No fertilizer, 1 = Fertilizer applied
rainfall = np.random.uniform(50, 300, n)  # Rainfall (mm)
sunlight_hours = np.random.uniform(4, 12, n)  # Sunlight hours per day
nitrogen = np.random.uniform(10, 50, n)  # Nitrogen content in soil
soil_treatment = np.random.choice([0, 1], size=n)  # 0 = No treatment, 1 = Soil treated
crop_type = np.random.choice(['wheat', 'rice', 'corn'], size=n)  # Crop type
field_size = np.random.uniform(1000, 5000, n)  # Field size (sq.m)
weed_presence = np.random.choice([0, 1], size=n)  # 0 = No, 1 = Yes
pest_presence = np.random.choice([0, 1], size=n)  # 0 = No, 1 = Yes
irrigation = np.random.choice([0, 1], size=n)  # 0 = No irrigation, 1 = Irrigated
farm_id = np.random.choice(range(1, 51), size=n)  # 50 different farms or regions
farm_location = np.random.choice(['North', 'South', 'East', 'West'], size=n)  # Farm location

# Creating DataFrame with the simulated data
data = pd.DataFrame({
    'temperature': temperature,
    'humidity': humidity,
    'soil_ph': soil_ph,
    'fertilizer': fertilizer,
    'rainfall': rainfall,
    'sunlight_hours': sunlight_hours,
    'nitrogen': nitrogen,
    'soil_treatment': soil_treatment,
    'crop_type': crop_type,
    'field_size': field_size,
    'weed_presence': weed_presence,
    'pest_presence': pest_presence,
    'irrigation': irrigation,
    'farm_id': farm_id,
    'farm_location': farm_location
})

# Encode categorical variables (e.g., crop_type and farm_location)
data['crop_type'] = data['crop_type'].map({'wheat': 0, 'rice': 1, 'corn': 2})
data['farm_location'] = data['farm_location'].map({'North': 0, 'South': 1, 'East': 2, 'West': 3})

# Feature selection for clustering (numeric columns)
features = data[['temperature', 'humidity', 'soil_ph', 'fertilizer', 'rainfall', 'sunlight_hours', 'nitrogen', 
                 'soil_treatment', 'field_size', 'weed_presence', 'pest_presence', 'irrigation', 'crop_type', 'farm_location']]

# Standardizing the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Apply Agglomerative Clustering (Hierarchical Clustering)
clustering = AgglomerativeClustering(n_clusters=5)  # Let's cluster into 5 groups
data['cluster'] = clustering.fit_predict(features_scaled)

# 1. Dendrogram visualization
linked = linkage(features_scaled, 'ward')

plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Farm samples")
plt.ylabel("Euclidean distances")
plt.show()

# 2. Clustering visualization using PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(features_scaled)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=data['cluster'], palette="viridis", s=10)
plt.title("PCA Plot of Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()

# Summarize features by cluster
clusters = data.groupby('cluster').mean()


# 3. Features vs clusters
cluster_means = data.groupby('cluster').mean()
plt.figure(figsize=(12, 8))
sns.heatmap(cluster_means, annot=True, cmap='coolwarm')
plt.title("Mean Feature Values by Cluster")
plt.show()

# 4. Temperature by cluster
plt.figure(figsize=(8, 6))
sns.boxplot(x='cluster', y='temperature', data=data)
plt.title("Temperature by Cluster")
plt.show()

# 5. Nitrogen by cluster
plt.figure(figsize=(8, 6))
sns.boxplot(x='cluster', y='nitrogen', data=data)
plt.title("Nitrogen by Cluster")
plt.show()

# 6. First 4 features with clustering
sns.pairplot(data[['temperature', 'humidity', 'soil_ph', 'fertilizer', 'cluster']], hue='cluster', palette='viridis')
plt.suptitle("Pairplot of First 4 Features with Clustering", y=1.02)
plt.show()

# 7. Cluster distribution across different farm locations
plt.figure(figsize=(8, 6))
sns.countplot(x='farm_location', hue='cluster', data=data)
plt.title("Cluster Distribution across Farm Locations")
plt.show()

# 8. Cluster distribution by crop type
plt.figure(figsize=(8, 6))
sns.countplot(x='crop_type', hue='cluster', data=data)
plt.title("Cluster Distribution by Crop Type")
plt.show()

# 9. Correlation matrix for selected features
corr_matrix = features.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Features")
plt.show()

# 10. Distribution of field sizes by clusters
plt.figure(figsize=(8, 6))
sns.boxplot(x='cluster', y='field_size', data=data)
plt.title("Field Size Distribution by Cluster")
plt.show()

# Interpretation generation function
def interpret_cluster(cluster_idx, cluster_data):
    interpretation = []
    interpretation.append(f"\nCluster {cluster_idx + 1} (Cluster Summary):")
    interpretation.append(f"- Temperature: {cluster_data['temperature']:.2f}Â°C")
    interpretation.append(f"- Humidity: {cluster_data['humidity']:.2f}%")
    interpretation.append(f"- Soil pH: {cluster_data['soil_ph']:.2f}")
    interpretation.append(f"- Fertilizer Use: {cluster_data['fertilizer'] * 100:.0f}% farms use fertilizer")
    interpretation.append(f"- Rainfall: {cluster_data['rainfall']:.2f} mm")
    interpretation.append(f"- Sunlight Hours: {cluster_data['sunlight_hours']:.2f} hours/day")
    interpretation.append(f"- Nitrogen Content: {cluster_data['nitrogen']:.2f}")
    interpretation.append(f"- Field Size: {cluster_data['field_size']:.2f} sq.m")
    interpretation.append(f"- Weed Presence: {cluster_data['weed_presence'] * 100:.0f}% farms have weeds")
    interpretation.append(f"- Pest Presence: {cluster_data['pest_presence'] * 100:.0f}% farms have pests")
    interpretation.append(f"- Farm Location Distribution: {data[data['cluster'] == cluster_idx]['farm_location'].value_counts().to_dict()}")
    
    return '\n'.join(interpretation)

summary = []
for i in range(5):
    cluster_info = clusters.iloc[i]
    summary.append(interpret_cluster(i, cluster_info))

print("\nInterpretation Summary:")
for line in summary:
    print(line)

```
